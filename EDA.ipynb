```python
# Assuming 'wb_dataset' is your DataFrame
numerical_features = ['farm_size', 'hectare', 'total_loan_value', 'repayment_value',
                       'amount_repaid', 'debt_to_farm_size_ratio', 'avg_loan_repayment_rate',
                       'time_since_last_loan']

# Add the target variable to the list
numerical_features.append('fully_repaid_within_maturity')

# Select relevant columns for analysis
selected_data = wb_dataset[numerical_features]

# Set up the matplotlib figure
plt.figure(figsize=(15, 10))

# Create scatter plots for each numerical feature against the target variable
for i, feature in enumerate(numerical_features[:-1]):
    plt.subplot(3, 3, i + 1)
    sns.scatterplot(x=feature, y='fully_repaid_within_maturity', data=selected_data, alpha=0.5)
    plt.title(f'{feature} vs Fully Repaid')
    plt.xlabel(feature)
    plt.ylabel('Fully Repaid')

# Adjust layout
plt.tight_layout()
plt.show()
```
